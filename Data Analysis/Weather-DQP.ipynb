## Missing handling

```python
weather.isnull().sum()
```

```python
weather_missing  = 100 * (weather.isnull().sum()/weather.shape[0])
weather_missing
```

**Result**
- each field represent
    - ``wind_gust``: Wind gust. Units – default: meter/sec
    - ``rain_1h``: Rain volume for the last hour, mm
    - ``snow_1h``: Snow volume for the last hour, mm (in liquid state)
    - ``visibility``:  Average visibility, metres. The maximum value of the visibility is 10km
- ``wind_gust``, ``rain_1h`` and ``snow_1h``: replace NA with 0. Because the value of those columns indicate the possibility of wind, rain and snow
- ``visibility``: replace NA with the average visibility of one hour before and the last hour. 

```python
# 'wind_gust', 'rain_1h' and 'snow_1h': replace NA with 0
weather[['wind_gust', 'rain_1h', 'snow_1h']] = weather[['wind_gust', 'rain_1h', 'snow_1h']].fillna(0)
weather
```

```python
# 'visibility': replace NA with the average visibility of one hour before and the last hour
weather[weather['visibility'].isnull()]['dt']
```

```python
# 2018-03-05 03:00:00
weather.loc[weather['dt'] == '2018-03-05 03:00:00', 'visibility'] = weather[(weather['dt'] == '2018-03-05 02:00:00') | (weather['dt'] == '2018-03-05 04:00:00')]['visibility'].sum() / 2

# 2018-09-17 21:00:00 to 2018-09-18 03:00:00
weather.loc[(weather['dt'] >= '2018-09-17 21:00:00') & (weather['dt'] <= '2018-09-18 03:00:00'), 'visibility'] = weather[(weather['dt'] == '2018-09-17 20:00:00') | (weather['dt'] == '2018-09-18 04:00:00')]['visibility'].sum() / 2

# 2018-11-18 09:00:00
weather.loc[weather['dt'] == '2018-11-18 09:00:00', 'visibility'] = weather[(weather['dt'] == '2018-11-18 08:00:00') | (weather['dt'] == '2018-11-18 10:00:00')]['visibility'].sum() / 2

# 2018-11-18 18:00:00 to 2018-11-19 08:00:00
weather.loc[(weather['dt'] >= '2018-11-18 18:00:00') & (weather['dt'] <= '2018-11-19 08:00:00'), 'visibility'] = weather[(weather['dt'] == '2018-11-18 17:00:00') | (weather['dt'] == '2018-11-19 09:00:00')]['visibility'].sum() / 2

# 2018-11-19 12:00:00 to 2018-11-19 15:00:00
weather.loc[(weather['dt'] >= '2018-11-19 12:00:00') & (weather['dt'] <= '2018-11-19 15:00:00'), 'visibility'] = weather[(weather['dt'] == '2018-11-19 11:00:00') | (weather['dt'] == '2018-11-19 16:00:00')]['visibility'].sum() / 2

# 2018-11-20 03:00:00
weather.loc[weather['dt'] == '2018-11-20 03:00:00', 'visibility'] = weather[(weather['dt'] == '2018-11-20 02:00:00') | (weather['dt'] == '2018-11-20 04:00:00')]['visibility'].sum() / 2

# 2018-11-20 06:00:00
weather.loc[weather['dt'] == '2018-11-20 06:00:00', 'visibility'] = weather[(weather['dt'] == '2018-11-20 05:00:00') | (weather['dt'] == '2018-11-20 07:00:00')]['visibility'].sum() / 2

# 2018-11-21 00:00:00 to 2018-11-21 01:00:00
weather.loc[(weather['dt'] >= '2018-11-21 00:00:00') & (weather['dt'] <= '2018-11-21 01:00:00'), 'visibility'] = weather[(weather['dt'] == '2018-11-20 23:00:00') | (weather['dt'] == '2018-11-21 02:00:00')]['visibility'].sum() / 2

# 2018-11-21 06:00:00
weather.loc[weather['dt'] == '2018-11-21 06:00:00', 'visibility'] = weather[(weather['dt'] == '2018-11-21 05:00:00') | (weather['dt'] == '2018-11-21 07:00:00')]['visibility'].sum() / 2

# 2018-11-21 08:00:00
weather.loc[weather['dt'] == '2018-11-21 08:00:00', 'visibility'] = weather[(weather['dt'] == '2018-11-21 07:00:00') | (weather['dt'] == '2018-11-21 09:00:00')]['visibility'].sum() / 2

# 2018-11-21 10:00:00 to 2018-11-21 11:00:00
weather.loc[(weather['dt'] >= '2018-11-21 10:00:00') & (weather['dt'] <= '2018-11-21 11:00:00'), 'visibility'] = weather[(weather['dt'] == '2018-11-21 09:00:00') | (weather['dt'] == '2018-11-21 12:00:00')]['visibility'].sum() / 2

# 2018-12-11 06:00:00
weather.loc[weather['dt'] == '2018-12-11 06:00:00', 'visibility'] = weather[(weather['dt'] == '2018-12-11 05:00:00') | (weather['dt'] == '2018-12-11 07:00:00')]['visibility'].sum() / 2
```

### data quality plan for weather
| Variable Names     | Data Quality Issue                        | Handling Strategy          |
|--------------------|-------------------------------------------|----------------------------|
| dt                   | Rows are not in 2018                    | Drop rows (18126 rows affect) |

| dt_iso               | Duplicated with dt                      | Drop column |       
| timezone             | Only relevant to daylight saving time   | Drop column |     
| city_name            | Cardinality is 1                        | Drop column | 
| lat                  | Constant columns                        | Drop column |     
| lon                  | Constant columns                        | Drop column |     
| visibility           | Missing data | replace with the average visibility of the closest time (37 rows affect) | 
| temp_min             | Irrelevant                              | Drop column | 
| temp_max             | Irrelevant                              | Drop column |   
| sea_level            | Missing is 100%                         | Drop column |   
| grnd_level           | Missing is 100%                         | Drop column |

| rain_1h              | Missing is 83%                          | Replace with 0 |
