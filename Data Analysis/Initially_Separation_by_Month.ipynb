```python
import pandas as pd
```

```python
# Very large file - will take a long time
leavetimes = pd.read_csv('rt_leavetimes_DB_2018.txt', sep=";", header=0) 
```

```python
leavetimes.DAYOFSERVICE.value_counts()
```

```python
leavetimes_raw = leavetimes.copy()
```

```python
# convert feature 'DAYOFSERVICE' to datetime and store as a new feature 'DAYOFSERVICE(datetime)'
leavetimes['DAYOFSERVICE(datetime)'] = pd.to_datetime(leavetimes.DAYOFSERVICE, 
                                  format = '%d-%b-%y %H:%M:%S', 
                                  errors = 'coerce')

leavetimes.head()
```

```python
# Check if any data in 'DAYOFSERVICE' is not in the format '%d-%b-%y %H:%M:%S'
leavetimes['DAYOFSERVICE(datetime)'].isnull().sum()
```

```python
# create a new feature 'MONTH'
leavetimes['MONTH'] = leavetimes['DAYOFSERVICE(datetime)'].dt.month

leavetimes.head()
```

```python
leavetimes['MONTH'].value_counts()
```

```python
# breakdown of the data by month
for i in range(1, 13):
    save_data = leavetimes[leavetimes['MONTH'] == i]
    file_name = 'month_' + str(i) + '.csv'
    save_data.to_csv(file_name, index=False)  
```
