```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Final_DB/trips_journey_time_cleaned.csv')
df.head()
```

```python
lineids = list(df['LINEID'].unique())
```

```python
len(lineids)
```

### result of train dataset

```python
result_train = pd.DataFrame(columns=['stop', 'mae', 'mape', 'mse', 'rmse', 'r2', 'lineid'])
result_train
```

```python
for lineid in lineids:
    print(lineid)
    df1 = pd.read_csv(f'result/{lineid}_train.csv')
    result_train = pd.concat([result_train, df1], ignore_index=True)
```

```python
result_train
```

```python
result_train.dtypes
```

```python
for numeric_column in ['mae', 'mape', 'mse', 'rmse', 'r2']:
    result_train[numeric_column] = result_train[numeric_column].astype('float')

result_train.dtypes
```

```python
result_train.describe().T
```
