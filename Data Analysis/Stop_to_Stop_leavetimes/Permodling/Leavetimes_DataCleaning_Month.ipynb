### 4. check whether the value of ``PROGRNUMBER`` is sequential

```python
leavetimes = leavetimes.sort_values(by=['DAYOFSERVICE', 'TRIPID','PROGRNUMBER'], axis=0, ascending=True)
leavetimes.reset_index(drop=True, inplace=True)
leavetimes.head()
```

```python
leavetimes['SKIP'] = np.nan
```

```python
leavetimes.loc[(leavetimes['DAYOFSERVICE'] == leavetimes['DAYOFSERVICE'].shift(-1)) & 
               (leavetimes['TRIPID'] == leavetimes['TRIPID'].shift(-1)), 'SKIP'] = leavetimes['PROGRNUMBER'].shift(-1)

leavetimes.head()
```

```python
leavetimes['SKIP'] = leavetimes['SKIP'] - leavetimes['PROGRNUMBER']

leavetimes.head()
```

```python
leavetimes['SKIP'].value_counts(dropna = True, normalize = True)
```

Result: 
- About 7% rows of the data is not sequential which means lots of trips have non-sequential data
- Leave as it

### 5. Check all trips achieved, the value of  ``SUPPRESSED``

```python
# If the leave time is planned and achieved, the record will contain both the planned and actual data.
# If the leave time is planned and suppressed, the record will contain only the planned data and the flag “Suppressed” will be checked
# When the trip is partially suppressed it says that the previous link is suppressed (0 = achieved, 1 = suppressed)

# If the trips is partially suppressed the flag “Suppressed” will not be checked. The set of the leave times suppressed is detailed in table RT_LeaveTimes

leavetimes['SUPPRESSED'].value_counts(dropna = False)
```

```python
leavetimes_partially_suppressed_achieved = leavetimes[leavetimes['SUPPRESSED'] == 0.0]
leavetimes_partially_suppressed_achieved
```

```python
# inspect 'SUPPRESSED' = 0(achieved) to see how 'SUPPRESSED' = 0.0 affect the data
leavetimes[(leavetimes['DAYOFSERVICE'] == '2018-01-01') & (leavetimes['TRIPID'] == 5963473)].sort_values(by='PROGRNUMBER', axis=0, ascending=True)
```

```python
# drop the trip of a certain day whose data of leavetime contains at least one row that is partially suppressed
leavetimes_partially_suppressed_suppressed = leavetimes[leavetimes['SUPPRESSED'] == 1.0]
leavetimes_partially_suppressed_suppressed
```

**Result**
- drop the trips which contains a row that is suppressed

```python
df1 = leavetimes_partially_suppressed_suppressed[['DAYOFSERVICE', 'TRIPID']]
df1.head()
```

```python
df2 = leavetimes.merge(df1, on = ['DAYOFSERVICE', 'TRIPID'])
df2.head()
```

```python
leavetimes = pd.concat([leavetimes,df2]).drop_duplicates(keep=False)
leavetimes
```

### 6. check irrelevant columns

```python
# After delete partially suppressed trips, ``SUPPRESSED`` and ``JUSTIFICATIONID`` are irrelevant
# drop 'SUPPRESSED' and 'JUSTIFICATIONID'
leavetimes = leavetimes.drop(['SUPPRESSED', 'JUSTIFICATIONID'], axis=1)

leavetimes.head()
```

```python
# drop 'LASTUPDATE' which is irrelevant
leavetimes = leavetimes.drop(['LASTUPDATE'], axis=1)

leavetimes.head()
```

```python
leavetimes_times_24h = leavetimes[(leavetimes['ACTUALTIME_ARR'] > 86400) | (leavetimes['ACTUALTIME_DEP'] > 86400) |
                                 (leavetimes['PLANNEDTIME'] > 86400)] 

leavetimes_times_24h
```

Result: 
- Leave as it. 
- AVL data refer to one day of service that could be greater than 24 hours
- convert 'PLANNEDTIME', 'ACTUALTIME_ARR' and 'ACTUALTIME_DEP' to datetime

```python
# convert 'PLANNEDTIME', 'ACTUALTIME_ARR' and 'ACTUALTIME_DEP' to datetime
leavetimes['PLANNEDTIME'] = leavetimes['DAYOFSERVICE'] + pd.to_timedelta(leavetimes['PLANNEDTIME'], unit='S')
leavetimes['ACTUALTIME_ARR'] = leavetimes['DAYOFSERVICE'] + pd.to_timedelta(leavetimes['ACTUALTIME_ARR'], unit='S')
leavetimes['ACTUALTIME_DEP'] = leavetimes['DAYOFSERVICE'] + pd.to_timedelta(leavetimes['ACTUALTIME_DEP'], unit='S')

leavetimes
```

### data quality plan for leavetimes
| Variable Names     | Data Quality Issue                        | Handling Strategy          |
|--------------------|-------------------------------------------|----------------------------|
| DATASOURCE         | Irrelevant                                | ---                        |
| DATASOURCE         | Cardinality is 1                          | Drop column                |
| PROGRNUMBER        | About 7% rows of the data is not sequential | Leave as it |
| PLANNEDTIME_ARR, PLANNEDTIME_DEP    | Duplicated columns | Store value in a new feature 'PLANNEDTIME'|
| ACTUALTIME_ARR, ACTUALTIME_DEP      | ACTUALTIME_ARR > ACTUALTIME_DEP | Drop row (1 row affected))
| PASSENGERS         | Missing is 100%                           | Drop column                |
| PASSENGERSIN       | Missing is 100%                           | Drop column                |
| PASSENGERSOUT      | Missing is 100%                           | Drop column                |
| DISTANCE           | Missing is 100%                           | Drop column                |
| SUPPRESSED         | Partially suppressed (0 = achieved, 1 = suppressed) | Drop rows (1 = suppressed) |
| SUPPRESSED         | Irrelevant after drop, all rows contain achieved time | Drop column.   |        
| JUSTIFICATIONID    | Irrelevant, and the existence of the value corresponds synchronously with the existence of the value of 'SUPPRESSED' | Drop column |
| LASTUPDATE         | Irrelevant                                | Drop column                |
| NOTE               | Missing is 100%                           | Drop column                |

## CONCLUSION
- Each record of this table contains the actual data for one leave time of one vehicle at one stop point of the route.
- When the trip is partially suppressed it says that the previous link is suppressed (0 = achieved, 1 = suppressed)
    - If the leave time is planned and achieved, the record will contain both the planned and actual data.
    - If the leave time is planned and suppressed, the record will contain only the planned data and the flag “Suppressed” will be checked.
- Primary KEY: DayofService, TripID, ProgrNumber
- Foreign KEY: TripID, StopPointID, VehicleID
