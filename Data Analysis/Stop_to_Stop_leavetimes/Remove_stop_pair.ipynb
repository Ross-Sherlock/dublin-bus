```python
df = pd.read_csv('../Final_DB/trips_journey_time_cleaned.csv')
df.head()
```

```python
lineids = list(df['LINEID'].unique())
```

```python
len(lineids)
```

# read 

```python
remove = pd.DataFrame(columns=['DAYOFSERVICE', 'STARTSTOP', 'LINEID', 'HOUR', 'ENDSTOP', 'JOURNEYTIME', 'PLANNED_JOURNEYTIME', 'STOP_TIME'])
remove
```

```python
for lineid in lineids:
    print(lineid)
    
    df1 = pd.read_csv(f'result/{lineid}_remove.csv')
    df2 = pd.read_csv(f'../Final_DB/Lines/{lineid}/{lineid}.csv')
    df1 = df1.merge(df2, on = ['STARTSTOP', 'ENDSTOP'])
    
    remove = pd.concat([remove, df1], ignore_index=True)
```

```python
remove.head()
```

```python
remove_result = remove.groupby(['LINEID', 'STARTSTOP', 'ENDSTOP']).median()
remove_result
```

```python
remove_result = remove_result.drop(['HOUR', 'JOURNEYTIME', 'STOP_TIME'], axis = 1).reset_index()
remove_result.head()
```

```python
remove_result.to_csv(f'remove_prediction.csv', index = False)
```
